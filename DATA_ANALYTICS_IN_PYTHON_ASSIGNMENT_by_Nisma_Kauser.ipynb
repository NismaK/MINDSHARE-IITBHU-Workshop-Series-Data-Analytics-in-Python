### **1. DATA COLLECTION:** Importing and exploring the datasets.

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
df = pd.read_csv("Dataset.csv")
#Display the first 5 rows of dataset
df.head()
```

```python
#Display the last 5 rows of dataset
df.tail()
```

```python
df.shape#print shape(row,column)
```

### **2. DATA CLEANING:** Handling missing values and incorrect data.

```python
#Check for missing values
df.isnull()
```

```python
df.isnull().sum()
```

```python
#Removing if any null value present
df.dropna(inplace=True)
```

```python
#Discovering duplicates
print(df.duplicated())
```

```python
#Removing duplicates
df.drop_duplicates(inplace=True)
```

### **3. DATA ANALYSIS:**  Analyzing trends, patterns, and relationships.

```python
df.describe()
```

```python
df.info()#checks what kind of datas are present
```

```python
# Analyze trends, e.g., number of tourists over time
tourists_trend = df.groupby('Year')['Number_of_Tourists'].sum()
print("\nNumber of tourists over time:")
print(tourists_trend)
```

```python
# Analyze relationships, e.g., tourists vs. revenue
correlation = df['Number_of_Tourists'].corr(df['Revenue'])
print("\nCorrelation between number of tourists and revenue:", correlation)
```

```python
sns.boxplot(data=df, x='Number_of_Tourists', y='Revenue')
plt.title("Relationship between Tourists and Revenue")
plt.show()
```

```python
# Analyze the average spending per tourist over the years for each country
df['Spending_per_Tourist'] = df['Revenue'] / df['Number_of_Tourists']
average_spending = df.groupby(['Year', 'Country'])['Spending_per_Tourist'].mean()
print("\nAverage spending per tourist over the years for each country:")
print(average_spending)
```

```python
# Analyzing the total revenue generated by each country over the given period
total_revenue = df.groupby('Country')['Revenue'].sum()
print("\nTotal revenue generated by each country:")
print(total_revenue)
```

```python
# Analyzing the yearly change in the number of tourists for each country
df['Tourists_Change'] = df.groupby('Country')['Number_of_Tourists'].diff()
print("\nYearly change in the number of tourists for each country:")
print(df[['Year', 'Country', 'Tourists_Change']].dropna())
```

### **4.DATA VISUALIZATION:**  Creating visual representations of the data.

```python
# Group data by Year and sum the number of tourists
tourists_trend = df.groupby('Year')['Number_of_Tourists'].sum().reset_index()

# Plot the number of tourists over time using a bar chart
plt.figure(figsize=(10, 6))
sns.barplot(data=tourists_trend, x='Year', y='Number_of_Tourists', palette='rainbow',hue='Number_of_Tourists')
plt.title('Number of Tourists Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Tourists')
plt.show()
```

```python
# Plot the number of tourists over time
plt.figure(figsize=(8, 8))
sns.lineplot(data=df, x='Year', y='Number_of_Tourists', hue='Country')
plt.title('Number of Tourists Over Time')
plt.xlabel('Year')
plt.ylabel('Number of Tourists')
plt.legend(title='Country')
plt.show()
```

```python
# Plot revenue distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Country', y='Revenue',palette='pastel',hue='Country')
plt.title('Revenue Distribution by Country')
plt.xlabel('Country')
plt.ylabel('Revenue')
plt.show()
```

```python
#Distribution of Tourists by Country
tourists_by_country = df.groupby('Country')['Number_of_Tourists'].sum()
plt.figure(figsize=(10, 8))
plt.pie(tourists_by_country, labels=tourists_by_country.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Wistia', len(tourists_by_country)),shadow=True,)
plt.title('Distribution of Tourists by Country')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
